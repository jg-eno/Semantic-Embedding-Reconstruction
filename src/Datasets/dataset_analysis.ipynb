```python
from datasets import load_dataset

ds = load_dataset("microsoft/ms_marco","v1.1",split="validation")
```

```python
print(ds)
```

```python
len(ds)
```

```python
ds[0]
```

```python
pg = ds[0]['passages']['passage_text']
```

```python
len(pg)
```

```python
from transformers import AutoTokenizer
MODEL_NAME = "Qwen/Qwen3-Embedding-0.6B"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
```

```python
MAX_LENGTH = 512
inputs = tokenizer(
                pg[0],
                return_tensors="pt",
                padding=True,
                truncation=True,
                max_length=MAX_LENGTH
            )
print(len(inputs['input_ids'][0]))
```

```python
total = 0
for example in ds:
    for paragraph in example["passages"]["passage_text"]:
        inputs = tokenizer(
                paragraph,
                return_tensors="pt",
                padding=True,
                truncation=True,
                max_length=MAX_LENGTH
            )
        print(len(inputs['input_ids'][0]))
        total += len(inputs['input_ids'][0])
        break
```

```python
print(total/len(ds))
```
